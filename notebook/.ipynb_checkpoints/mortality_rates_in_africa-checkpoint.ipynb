#### AFRICAN COUNTRIES BY INFANT MORTALITY RATE

```python
df_imc = df.groupby('Entity', as_index=False)['Infant Mortality rate'].sum()
df_imc_sorted = df_imc.sort_values(by='Infant Mortality rate', ascending=False)
df_imc_sorted.head()
```

```python
# Limit to the first 7 rows only
df_imc_sorted = df_imc_sorted.head(7)

# Normalize the values to be used as proportions for the area
df_imc_sorted['Normalized'] = df_imc_sorted['Infant Mortality rate'] / df_imc_sorted['Infant Mortality rate'].sum()

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Define a colormap with distinct colors
cmap = plt.get_cmap('tab10')  

# Set initial position
x_start = 0
y_start = 0
height = 1 

# Loop through each rectangle
for idx, (row_index, row) in enumerate(df_imc_sorted.iterrows()):
    width = row['Normalized'] 
    color = cmap(idx / len(df_imc_sorted))  
    ax.add_patch(plt.Rectangle((x_start, y_start), width, height, color=color, label=row['Entity']))
    x_start += width  

# Add category labels
x_start = 0  
for idx, row in df_imc_sorted.iterrows():
    ax.text(x_start + row['Normalized'] / 2, 0.5, row['Entity'], ha='center', va='center', 
            color='white', fontweight='bold', rotation=90)  
    x_start += row['Normalized']  

# Customize the plot
ax.set_title('African countries by Infant mortality rate')
ax.axis('off') 

# Add legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc='upper left', title='Entities')

# Show the plot
plt.show()
```

#### CORRELATION OF INFANT IMMUNIZATION INDICATORS

```python
# Create a list of Country, Year, and Infant Immunization rates
df_vac = df.columns[df.columns.str.endswith(('ized)', 'tity', 'ear'))].tolist()
df_vac
```

```python
# Create a dataframe for Infant Immunization rates
dn = df[df_vac]
coln = df_vac[2:]
dn = dn.dropna(subset=coln, how='all').reset_index(drop=True)
dn.head()
```

```python
# Create a Correlation matrix of numeric columns
dc = dn[df_vac[2:]]
correlation_matrix = dc.corr().round(2)
correlation_matrix
```

```python
# Correlation Heatmap of Infant immunization rates
plt.figure(figsize=(10, 8)) 

sns.heatmap(correlation_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            vmin=-1,
            vmax=1)

plt.title('Correlation Heatmap of Vaccination Rates')

plt.show()
```

```python
# Pairplots of infant immunization rates
sns.pairplot(dc)
plt.show()
```

#### DISTRIBUTION OF INFANT IMMUNIZATION INDICATORS ACROSS AFRICAN COUNTRIES

```python
# List of columns to be aggregated
dn_cols = [
    'BCG (% of one-year-olds immunized)',
    'HepB3 (% of one-year-olds immunized)',
    'Hib3 (% of one-year-olds immunized)',
    'IPV1 (% of one-year-olds immunized)',
    'MCV1 (% of one-year-olds immunized)',
    'PCV3 (% of one-year-olds immunized)',
    'Pol3 (% of one-year-olds immunized)',
    'RCV1 (% of one-year-olds immunized)',
    'RotaC (% of one-year-olds immunized)',
    'YFV (% of one-year-olds immunized)',
    'DTP3 (% of one-year-olds immunized)'
]

# Grouping and aggregating the data by 'Entity'
dn_grouped = dn.groupby('Entity', as_index=False)[dn_cols].agg('mean').round(2)

# Sorting by all columns in descending order
dn_grp_sorted = dn_grouped.sort_values(by=dn_cols, ascending=[False] * len(dn_cols))

# Set 'Entity' as the index
dn_grp_sorted.set_index('Entity', inplace=True)

# Calculate the total height of each bar (sum of all immunization rates for each Entity)
dn_grp_sorted['Total'] = dn_grp_sorted.sum(axis=1)

# Sort by the total height (sum of all columns) in descending order
dn_grp_sorted_sorted_by_total = dn_grp_sorted.sort_values(by='Total', ascending=False)

# Drop the 'Total' column for the plot, as it's only used for sorting
dn_grp_sorted_sorted_by_total.drop(columns='Total', inplace=True)

# Plot the stacked column chart
plt.figure(figsize=(12, 8))
dn_grp_sorted_sorted_by_total.plot(kind='bar', stacked=True, figsize=(12, 8))

# Customize the plot
plt.title('Average Immunization Rates by Entity (Sorted by Total Immunization)', fontsize=16)
plt.xlabel('Entity', fontsize=14)
plt.ylabel('Immunization Rate (%)', fontsize=14)
plt.xticks(rotation=90)
plt.legend(title='Immunization Types', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot with tight layout
plt.tight_layout()
plt.show()
```
