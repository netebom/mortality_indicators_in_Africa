```python
dfs = [df1, df2, df4, df5, df6, df7, df8, dt]
dn = dfs[0]

for df in dfs[1:]:
    dn = pd.merge(dn, df, on=['Entity', 'Code', 'Year'], how='left')

```

```python
dn = dn.drop(columns='959828-annotations')
dn.head()
```

```python
dn.rename(
    columns={
        'Observation value - Indicator: Under-five mortality rate - Sex: Total - Wealth quintile: Total - Unit of measure: Deaths per 100 live births': 'Under five mortality rate (per 100 live births)'
    }, inplace=True
)
dn.info()
```

```python
dn['Cause of Death'] = dn['Cause of Death'].fillna('Not recorded')
```

```python
dn_u5 = dn.groupby('Entity', as_index=False)['Under five mortality rate (per 100 live births)'].sum()
dn_u5_sorted = dn_u5.sort_values(by='Under five mortality rate (per 100 live births)', ascending=False) 
dn_u5_sorted.head()
```

In Africa, the top 5 countries with high mortality rate of children

```python
dn_y5 = dn.groupby('Year', as_index=False)['Under five mortality rate (per 100 live births)'].sum()
dn_y5
dn_y5_sorted = dn_y5.sort_values(by='Under five mortality rate (per 100 live births)', ascending=False) 
dn_y5_sorted.head(3)
```
